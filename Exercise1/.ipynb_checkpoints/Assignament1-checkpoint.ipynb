```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image

text = open("my_keywords.txt").read()

print(text)
stopwords = set(STOPWORDS)

stopwords.update(["and", "the", "of", "to"])

wordcloud = WordCloud(
    stopwords=stopwords,
    max_font_size=70,
    background_color="white",
    width=600,
    height=400
).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file("my_wordcloud.png")
```
